### User guide
##### *Note: only one list (pulse, isi, or iti) can have multiple elements in a given plot

_pulse_: duration(s) of each stimulus (in seconds) 

_isi_: time between stimuli (in seconds)  

_iti_: time between trials (in hours)

_n_: number of pulses in trial

_statistic_: what statistic to sample with the bootstrap (default is median)

_integrate_: How long after pulse to integrate for total response (in seconds)


```python
#constant pulse, vary ISI
pulse=1
isi=[15,30,60,]
n=20

# # constant ISI, vary pulse
# pulse=[.5,1,2,5]
# isi=[30,]
# n=20

# pulse=[1,2,5]
# isi=[60,]
# n=10
# isi.reverse()
fig,ax = pulseTrain(pulse=pulse,isi=isi,n_boot=1e1,norm_time=0,trial_rng=(0,24),n=n,
                      integrate=30,exclude=['regen','vibe'])
```
